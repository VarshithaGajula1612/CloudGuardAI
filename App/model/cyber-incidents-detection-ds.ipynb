Dataset link: https://www.kaggle.com/datasets/mustafahabeeb90/cybersecurity-incident-dataset

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/cybersecurity-incident-dataset/cybersecurity synthesized data.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
plt.figure()
df['attack_type'].value_counts().head(10).plot(kind='bar', color='steelblue')
plt.title('Top 10 Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
df['outcome'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Attack Outcomes')
plt.ylabel('')
plt.show()
```

```python
plt.figure()
df['attack_severity'].plot(kind='hist', bins=10)
plt.title('Distribution of Attack Severity')
plt.xlabel('Severity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
df['target_system'].value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Targeted Systems')
plt.xlabel('Count')
plt.show()
```

```python
plt.figure()
df['industry'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Targeted Industries')
plt.xlabel('Industry')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure()
df.groupby('attack_type')['attack_severity'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Attack Severity by Attack Type')
plt.ylabel('Mean Severity')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['data_compromised_GB'])
plt.title('Distribution of Data Compromised (GB)')
plt.xlabel('GB')
plt.show()
```

```python
plt.figure()
df['response_time_min'].plot(kind='hist', bins=20)
plt.title('Distribution of Response Time (minutes)')
plt.xlabel('Response Time (min)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['data_compromised_GB', 'attack_duration_min', 'attack_severity', 'response_time_min']].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure()
df.set_index('timestamp').resample('M').size().plot()
plt.title('Monthly Trend of Cyberattacks')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='attack_duration_min', y='attack_severity', data=df, alpha=0.4)
plt.title('Attack Duration vs Attack Severity')
plt.xlabel('Duration (min)')
plt.ylabel('Severity')
plt.show()
```

```python
plt.figure()
df.groupby('attack_type')['response_time_min'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Response Time by Attack Type')
plt.ylabel('Mean Response Time (min)')
plt.show()
```

```python
plt.figure()
df['security_tools_used'].value_counts().head(10).plot(kind='barh')
plt.title('Most Common Security Tools Used')
plt.xlabel('Frequency')
plt.show()
```

```python
plt.figure()
df['mitigation_method'].value_counts().head(10).plot(kind='bar')
plt.title('Most Common Mitigation Methods')
plt.xlabel('Mitigation Method')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
df.groupby('attack_type')['data_compromised_GB'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Data Compromised by Attack Type')
plt.ylabel('Mean GB Compromised')
plt.show()
```

```python
plt.figure()
df['location'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Attacker Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
df['user_role'].value_counts().head(10).plot(kind='bar')
plt.title('Distribution of User Roles in Attacks')
plt.xlabel('User Role')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='attack_duration_min', y='response_time_min', data=df, alpha=0.3)
plt.title('Attack Duration vs Response Time')
plt.xlabel('Duration (min)')
plt.ylabel('Response Time (min)')
plt.show()
```

```python
sns.pairplot(df[['data_compromised_GB', 'attack_duration_min', 'attack_severity', 'response_time_min']])
plt.suptitle('Pairwise Relationships Between Key Numeric Features', y=1.02)
plt.show()
```

```python
df['hour'] = df['timestamp'].dt.hour
plt.figure()
df['hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Attack Frequency by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Attacks')
plt.show()
```
